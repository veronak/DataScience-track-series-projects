```python
red_wine_data.corr(method='pearson')
```

```python
#set width and height of the figure
plt.figure(figsize=(14,7))
#title
plt.title("Correlation Map for Red Wine")
#Heat map showing correlation in data
sns.heatmap(data=(red_wine_data.corr(method='pearson')), annot=True)
```

Positive Correlation for Red wine
There is a strong positive correlation between;
    fixed acidity and density
    fixed acidity and citric acid
    free sulfurdioxide and total sulfurdioxide
    alcohol and quality
Negative Correlation for Red wine
There is a high negative correlation between;
    density and alcohol
    pH and fixed acidity
    citric acid and pH
    volatile acidity and citric acid


```python
#set width and height of the figure
plt.figure(figsize=(14,7))
#title
plt.title("Correlation Map for White Wine ")
#Heat map showing correlation in data
sns.heatmap(data=(white_wine_data.corr(method='pearson')), annot=True)
```

Positive correlation in White Wine
There is a high positive correlatation between.
    density and residual sugar
    free sulfurdioxide and total sulfurdioxide
Negative correlation
There is almost no relationship between;
    citric acid, free sulfurdioxide and quality
